```python
import numpy as np
import pandas as pd
```

```python
train_df = pd.read_csv('./input/train.csv')
test_df = pd.read_csv('./input/test.csv')

full_df = pd.concat([train_df, test_df], ignore_index=True)
```

```python
full_df
```

#### 時系列特徴量

```python
full_df['datetime'] = pd.to_datetime(full_df['datetime'])
```

```python
# ラグ特徴量・ローリング特徴量を作成する対象列
target_col = 'y'

# ラグ作成（過去1日、2日、3日分など）
for lag in [1, 2, 3, 4, 5]:
    full_df[f'{target_col}_{lag}lag'] = full_df[target_col].shift(lag)
```

```python
for window in [3, 7, 14]:
    full_df[f'{target_col}_rolling_mean_{window}'] = full_df[target_col].shift(1).rolling(window=window).mean()
    full_df[f'{target_col}_rolling_max_{window}'] = full_df[target_col].shift(1).rolling(window=window).max()
    full_df[f'{target_col}_rolling_min_{window}'] = full_df[target_col].shift(1).rolling(window=window).min()
```

```python
train_features = full_df.iloc[:len(train_df)]
test_features = full_df.iloc[len(train_df):]
```

#### 前処理済みのファイルをアウトプット

```python
train_features.to_csv('./output/中間データ/train_with_時系列特徴量.csv', index=False)
test_features.to_csv('./output/中間データ/test_with_時系列特徴量.csv', index=False)
```
