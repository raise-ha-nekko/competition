## 2-1. 相関関係から因果関係へ

#### 因果関係の基準
ヒュームの３条件
 - 交絡因子（原因と結果の両方に影響する因子）が存在しない
 - 原因は結果に時間的に先行している
 - 原因と結果に相関関係がある

Hillの基準
 - 相関関係の強さ
 - 相関関係の一致性：相関関係の大きさはさまざまな状況で、対象や実証に利用する手法が違っていても一致している
 - 相関関係の特異性：Bと「A以外に原因として想定される変数」の相関は高くない、またAと「B以外の結果変数」の相関も高くない
 - 時間的な先行性
 - 量と反応関係の成立：原因となる変数Aが大きくなると、単調に結果となる変数Bの値も大きくなる
 - 妥当性
 - 先行知見との整合性
 - 実験による知見
 - 他の知見との類似性

 →重要なのは、因果関係を特定するために、**プロセスやメカニズムと統計的関連性**から考える必要があるということ。

## 2-2. ランダム化比較試験とは

#### エビデンスレベル
 - レベル1：処置群の前後比較、または処置群と対照群の処置後の1時点データのみを使用した分析（変数調整なし）
 - レベル2：レベル1において変数調整をした分析
 - レベル3：処置群と対照群それぞれの処置前後のデータを使用した分析（DiDなど）
 - レベル4：実験に類似する状況を利用した、回帰不連続デザインや操作変数法による分析
 - レベル5：処置群と対照群のランダム割り当てによる分析　←**ランダム化比較試験(RCT)**

#### RCTの問題点
 1. 倫理的にRCTができない場合が存在する
 2. 対象者が実験方法に従わない場合が存在する
 3. サンプルサイズが小さい場合に実施が困難
 4. 実施に費用と労力がかかる

## 2-3. 因果推論

#### 因果推論の対象範囲
 - 観察された因果関係における因果効果はどれほどか？（因果効果の推定）
    - 因果関係の方向を既知として因果効果を推定
 - 観察対象のプロセスやメカニズムの因果構造はどのようになっているか？（因果探索）
    - 因果の方向と強さを推定

## 2-4. 因果効果


#### 因果効果の定義
 - 事実：処置Tが実行された場合(T=1)
 - 反事実：処置Tが実行されなかった場合(T=0)  
の時、個体iの因果効果τ<sub>i</sub> = Y<sub>i</sub>(1) - Y<sub>i</sub>(0)

||処置|結果|
|-|-|-|
|実世界|T=1(処置あり)|Y<sub>i</sub>(1)|
|反事実世界|T=0(処置なし)|Y<sub>i</sub>(0)|

{Y<sub>i</sub>(1), Y<sub>i</sub>(0)}：潜在的結果変数  
とはいえ、潜在的結果変数を両方とも観測することは現実的には困難  
→ 推定する因果効果の対象を個体から集団に拡げることで因果効果を推定する。(RCT)

#### 因果効果の種類
 - ATE（Average Treatment Effect）: 平均処置効果
    - 集団全体に対して一部に処置群があり、その他は対照群となる場合、全体に対して処置をした場合の平均的な因果効果
    - E[Y(1)] - E[Y(0)]
    - 例）集団全体に対するワクチン接種の効果検証
 - ATT(Average Treatment effect on the Treated): 処置群への平均処置効果
    - 処置群のみに対する因果効果
    - 例）副反応の効果検証

 - CATE(Conditional Average Treatment Effect): 条件付き平均処置効果
    - ATEを特定の条件付けをして算出したもの、条件付けをすることでサブグループに分かれ、そのサブグループ内での因果効果を算出できる
 - ITE(Individual Treatment Effect): 個別の因果効果
    - 個体一つひとつの因果効果

→ ATE > CATE > ITEへと進むにつれて、よ理個別かされた処置効果の推定が可能になる。

#### 因果推論の重要な仮定：識別仮定(識別性条件 + SUTVA)
 - 識別性：観測データから母数が一意に推定できること
    - 独立性：処置あり/なしの場合に限らず、結果Yは独立する
    - 正値性：各処置に割り当てられる確率が0でない
 - SUTVA(Stable Unit Treatment Value Assumption)：処置を受ける個体において処置の効果が安定しているという仮定
    - 相互作用なし：潜在的結果変数が他の個体の処置状態に依存しない
    - 一致性：処置の結果が潜在的な結果と一致する

 1. 独立性(交換可能性)
    - ${Y(1), Y(0)} \perp T$
    - 独立性を仮定すると、処置の有無に関わらず潜在的結果変数が同じであるため、２群間の潜在的結果変数の差は処置Tによる因果効果と解釈できる
    - 特定の条件を伴って使用されることも多い。(条件付き独立性)
      - 共変量Xに対して、${Y(1), Y(0)} \perp T|X$
 2. 正値性
    - 処置群に割り当てられる確率も対照群に割り当てられる確率も0ではない状態
    - $0 < P(T=1) < 1$
 3. 相互作用なし
    - 1人が何らかの処置を受けても、それが他の人に影響を及ぼさない状況
    - Y<sub>i</sub> = T<sub>i</sub>Y<sub>i</sub>(1) + (1 - T<sub>i</sub>)Y<sub>i</sub>(0)
 4. 一致性
    - 処置の定義が不明確な場合にはこの条件は満たされない

→ 観測データの場合には処置がランダムに割り当てられていないため、4つの仮定の確認から始める。
→ その際に、データの背景を理解するため、グラフを用いて視覚的に因果の方向を考える。

## 2-5. 因果効果推定のためのグラフ

#### シンプソンのパラドックス
 - サンプル全体における相関とサンプルを部分的に取り出した際の部分ごとの相関が異なる現象

#### 交絡因子・共変量
 - 交絡因子：背景因子として原因にも結果にも影響を与えているもの
    - 対策：原則としては交絡因子を調整し、原因と結果の関係のみを取り出す
 - 共変量：結果変数に影響を与えるが、処置の影響を受けない変数
 - 中間変数：結果変数に影響する変数の中で、処置変数の影響を受けている変数

#### グラフィカルモデルを用いた因果の表現
 - グラフィカルモデル：ノードとエッジで構成されたもの
    - 無向グラフ
    - 有向グラフ
        - **非巡回的グラフ(DAG)**
            - 変数間の相互関係を可視化できる
            - 各変数の発生確率(同時確率分布)を条件付き確率の掛け合わせとして要素分解できる
        - 巡回的グラフ
